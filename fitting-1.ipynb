```julia
function abs_diff(x, mean)
    return sum(abs.(x .- mean))
end
```

```julia
function line_abs_diff(x, b, m, y)
    y_pred = b .+ x .* m
    return abs_diff(y_pred, y)
end
```

```julia
function objective_abs(params)
    b = params[1]
    m = params[2]
    @show b, m # this macro is very useful for doing 
    return line_abs_diff(data.x, b, m, data.y) # the output for this function
    # to be a single valued in order for it to be a single data
end
```

```julia
result = optimize(objective_abs, [b,m] .+ 0.) # the floating point array also works for this problem
# here Justin did type casting to float64
# the second argument is for the starting parameters
```

```julia
b_abs,m_abs = Optim.minimizer(result) # this function will output the optimized value
```

```julia
f = Figure()
Axis(f[1, 1])
errorbars!(data.x, data.y, data.sigma_y)
scatter!(data.x, data.y, markersize = 10, color = :black)
lines!(xx, b_abs .+ m_abs .* xx)
f
```

```julia
function gaussian_log_likelihood(x, mean, sigma)
    return sum(-log.(sigma * sqrt(2*π)) - 0.5 * (x - mean).^2 ./ sigma.^2)
end
```

```julia
gaussian_log_likelihood(data.y, m .* data.x .+ b, data.sigma_y)
```

```julia
function objective(params)
    b = params[1]
    m = params[2]
    y_pred = b .+ m .* data.x
    return -gaussian_log_likelihood(data.y .+ 0., y_pred, data.sigma_y .+ 0.)
end
```

```julia
[b,m].+0.
```

```julia
result = optimize(objective, [b,m] .+ 0.)
```

```julia
b_opt,m_opt = Optim.minimizer(result)
```

```julia
function objective_2(x, b, m, y, sigma_y)
    y_pred = b .+ m .* x
    return -gaussian_log_likelihood(y, y_pred, sigma_y)
end
```

```julia
result = optimize(p -> objective_2(data.x, p[1], p[2], data.y, data.sigma_y),
                  [b,m] .+ 0.);
Optim.converged(result)
```

```julia
b_opt,m_opt = Optim.minimizer(result)
```

```julia
f = Figure()
Axis(f[1, 1])
errorbars!(data.x, data.y, data.sigma_y)
scatter!(data.x, data.y, markersize = 10, color = :black)
lines!(xx, b_opt .+ m_opt .* xx, )
lines!(xx, b_abs .+ m_abs .* xx, color=:red)
f
```

```julia
# This process is called the jackknife method
n = length(data.x)

B_jack = zeros(n)
M_jack = zeros(n)

for i in 1:n
    x_sub = append!(data.x[1:(i-1)], data.x[(i+1):n]) # deleteat!(data.x, i),  Erick with the magic !!
    y_sub = append!(data.y[1:(i-1)], data.y[(i+1):n])
    s_sub = append!(data.sigma_y[1:(i-1)], data.sigma_y[(i+1):n])
    result = optimize(p -> objective_2(x_sub, p[1], p[2], y_sub, s_sub),
                      [b,m] .+ 0.); # anonymous function definition 
                                    # this grabs the global function at the -> result
    @assert(Optim.converged(result)) #@assert wiil evaluate the result is true of false
    b,m = Optim.minimizer(result)
    B_jack[i] = b
    M_jack[i] = m
end
```

```julia
# starting_param = [b_eye, m_eye]
result = optimize(p -> objective_outliers( b_jack, m_jack, alldata.x, alldata.y, alldata.sigma_y),
                  [b,m] .+ 0.);
Optim.converged(result)
b_punk, m_punk = Optim.minimizer(result)
```

```julia
starting_params = [700 + 0., -0.5]
result = optimize(p -> objective_outliers(p, alldata.x, alldata.y, alldata.sigma_y),
                  starting_params)
@assert Optim.converged(result)
b_punk, m_punk = Optim.minimizer(result)
```
